### 3. Define Dataset

```python
class LaserAudioDataset(Dataset):
    def __init__(self, dataframe, audio_dir, transform=None, label_encoder=None, use_waveform=False):
        self.annotations = dataframe
        self.audio_dir = audio_dir
        self.transform = transform
        self.label_encoder = label_encoder
        self.use_waveform = use_waveform
        
        # Create a mapping from label names to integer indices using the label encoder
        if label_encoder:
            self.label_to_index = {label: idx for idx, label in enumerate(label_encoder.classes_)}
        else:
            self.label_to_index = {label: idx for idx, label in enumerate(self.annotations.iloc[:, 3].unique())}

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        audio_path = os.path.join(self.audio_dir, self.annotations.iloc[idx, 0])
        waveform, sample_rate = torchaudio.load(audio_path)
        
        label = self.annotations.iloc[idx, 3]  # Assuming the label is in the fourth column
        # Convert label from string to integer index
        label = self.label_to_index[label]
        
        if self.use_waveform:
            # Return the raw waveform
            return waveform, label
        elif self.transform:
            # Apply transformation
            waveform = self.transform(waveform)
        
        return waveform, label
```

### 4. Define transformations

```python
# Define transformations for Spectrogram, Mel-Spectrogram, and MFCCs
def get_spectrogram_transform():
    return transforms.Spectrogram(n_fft=1024)

def get_melspectrogram_transform(sample_rate=44100):
    return transforms.MelSpectrogram(sample_rate=sample_rate, n_mels=64, n_fft=1024, hop_length=256)

def get_mfcc_transform(sample_rate=44100):
    return transforms.MFCC(sample_rate=sample_rate, n_mfcc=13, melkwargs={'n_fft': 1024, 'n_mels': 64, 'hop_length': 256})

# Define a function to get the appropriate transform based on a key
def get_transform(transform_type, sample_rate=44100):
    if transform_type == 'spectrogram':
        return get_spectrogram_transform()
    elif transform_type == 'melspectrogram':
        return get_melspectrogram_transform(sample_rate)
    elif transform_type == 'mfcc':
        return get_mfcc_transform(sample_rate)
    else:
        raise ValueError(f"Unknown transform type: {transform_type}")
```

Augmentations

**audiomentation** Library:
- TimeMasking: Randomly masks a section of the spectrogram along the time dimension. This can help the model learn to be invariant to temporal distortions.
- FrequencyMasking: Randomly masks a section of the spectrogram along the frequency dimension. This helps the model learn to be invariant to frequency distortions.
- AddGaussianNoise: Adds random Gaussian noise to the audio signal, simulating a noisy environment.
- TimeStretch: Stretches or compresses the time dimension of the audio without changing the pitch. This simulates variations in speed.
- PitchShift: Shifts the pitch of the audio up or down without changing the speed. This can simulate different speakers or instruments.
- Shift: Shifts the audio signal in time, introducing small temporal offsets.

**torchaudio-augmentations** Library:
- Gain: Randomly adjusts the gain (volume) of the audio signal.
- PolarityInversion: Inverts the polarity of the waveform, which can help the model learn polarity-invariant features.
- HighPassFilter: Applies a high-pass filter to the audio signal, removing low-frequency components below the specified cutoff frequency.
- LowPassFilter: Applies a low-pass filter to the audio signal, removing high-frequency components above the specified cutoff frequency.

```python
# Define data transformations for train and validation in a dictionary format
def create_data_transforms(transform_type, sample_rate=44100):
    data_transforms = {
        'train': vision_transforms.Compose([
            get_transform(transform_type, sample_rate),
            transforms.AmplitudeToDB(),
            transforms.TimeMasking(time_mask_param=80),  # Randomly masks a section of the spectrogram in the time dimension
            transforms.FrequencyMasking(freq_mask_param=80),  # Randomly masks a section of the spectrogram in the frequency dimension
            # Add custom augmentations from audiomentations
            lambda x: audiomentations.Compose([
                AddGaussianNoise(min_amplitude=0.001, max_amplitude=0.015, p=0.5),  # Adds Gaussian noise
                TimeStretch(min_rate=0.8, max_rate=1.25, p=0.5),  # Stretches the time dimension
                PitchShift(min_semitones=-4, max_semitones=4, p=0.5),  # Shifts the pitch
                Shift(min_fraction=-0.5, max_fraction=0.5, p=0.5)  # Shifts the waveform in time
            ])(x),
            # Add custom augmentations from torchaudio-augmentations
            lambda x: Compose([
                RandomApply([Gain(min_gain=-15, max_gain=15)], p=0.5),
                RandomApply([PolarityInversion()], p=0.3),
                RandomApply([HighPassFilter(min_cutoff_freq=500, max_cutoff_freq=1000)], p=0.5),
                # RandomApply([LowPassFilter(min_cutoff_freq=2000, max_cutoff_freq=8000)], p=0.5)
            ])(x)
        ]),
        'val': vision_transforms.Compose([
            get_transform(transform_type, sample_rate),
            transforms.AmplitudeToDB()
        ]),
    }
    return data_transforms
```

## 5. Consolidating Datasets

```python
# Create consolidated datasets with different transforms
def create_all_datasets(dataframe, audio_dir, transform_types, label_encoder, sample_rate=44100):

    all_datasets = {}
    
    for transform_type in transform_types:
        data_transforms = create_data_transforms(transform_type, sample_rate)
        
        # Create the main dataset
        main_dataset = LaserAudioDataset(dataframe=dataframe, audio_dir=audio_dir, label_encoder=label_encoder)
        
        # Create subsets using the stored indices
        train_dataset = Subset(main_dataset, train_indices)
        val_dataset = Subset(main_dataset, val_indices)
        test_dataset = Subset(main_dataset, test_indices)
        
        # Apply transformations to each subset
        train_dataset.dataset.transform = data_transforms['train']
        val_dataset.dataset.transform = data_transforms['val']
        test_dataset.dataset.transform = data_transforms['val']
        
        # Store datasets in a dictionary
        all_datasets[transform_type] = {
            'train': train_dataset,
            'val': val_dataset,
            'test': test_dataset
        }
    
    return all_datasets
```

```python
def normalize_tensor(tensor, min_value=0.0, max_value=1.0):
    min_tensor = tensor.min()
    max_tensor = tensor.max()
    normalized_tensor = (tensor - min_tensor) / (max_tensor - min_tensor)
    normalized_tensor = normalized_tensor * (max_value - min_value) + min_value
    return normalized_tensor

def get_frequency_bin_centers(transform_type, sample_rate, n_fft=1024, n_mels=64, n_mfcc=13):
    if transform_type == 'spectrogram':
        # FFT bin frequencies, linear scale
        freqs = np.fft.rfftfreq(n_fft, 1.0 / sample_rate)
    elif transform_type == 'melspectrogram':
        # Mel filter bank center frequencies, logarithmic scale
        mel_transform = torchaudio.transforms.MelSpectrogram(sample_rate=sample_rate, n_fft=n_fft, n_mels=n_mels)
        mel_frequencies = mel_transform.mel_scale.fb.numpy()
        freqs = mel_frequencies @ np.linspace(0, sample_rate // 2, mel_transform.n_mels)
    elif transform_type == 'mfcc':
        # MFCC uses Mel frequencies, logarithmic scale
        freqs = np.linspace(0, sample_rate // 2, n_mfcc)
    else:
        freqs = np.arange(10)
    return freqs

def convert_to_log_scale(freqs):
```

```python
import torch
import torch.nn as nn
import torchvision.models as models
from torchvision.models import (
    VGG16_Weights, VGG19_Weights, ResNet18_Weights, ResNet34_Weights, ResNet50_Weights,
    DenseNet121_Weights, GoogLeNet_Weights, EfficientNet_B0_Weights, 
    ViT_B_16_Weights, RegNet_Y_400MF_Weights, ConvNeXt_Tiny_Weights
)
import torchopenl3

# Create model function
def create_model(model_name, num_classes, pretrained=True):
    weights = None
    if pretrained:
        weights = {
            'vgg16': VGG16_Weights.DEFAULT,
            'vgg19': VGG19_Weights.DEFAULT,
            'resnet18': ResNet18_Weights.DEFAULT,
            'resnet34': ResNet34_Weights.DEFAULT,
            'resnet50': ResNet50_Weights.DEFAULT,
            'densenet121': DenseNet121_Weights.DEFAULT,
            'googlenet': GoogLeNet_Weights.DEFAULT,
            'efficientnet_b0': EfficientNet_B0_Weights.DEFAULT,
            'vit_b_16': ViT_B_16_Weights.DEFAULT,
            'regnet_y_400mf': RegNet_Y_400MF_Weights.DEFAULT,
            'convnext_tiny': ConvNeXt_Tiny_Weights.DEFAULT
        }.get(model_name, None)
    
    if model_name == 'vgg16':
        model = models.vgg16(weights=weights)
        model.features[0] = nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1)
        model.classifier[6] = nn.Linear(model.classifier[6].in_features, num_classes)
    elif model_name == 'vgg19':
        model = models.vgg19(weights=weights)
        model.features[0] = nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1)
        model.classifier[6] = nn.Linear(model.classifier[6].in_features, num_classes)
    elif model_name == 'resnet18' or model_name == 'resnet34':
        model = models.resnet18(weights=weights) if model_name == 'resnet18' else models.resnet34(weights=weights)
        model.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        model.fc = nn.Linear(model.fc.in_features, num_classes)
    elif model_name == 'efficientnet_b0':
        model = models.efficientnet_b0(weights=weights)
        model.features[0][0] = nn.Conv2d(1, 32, kernel_size=3, stride=2, padding=1, bias=False)
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
    elif model_name == 'vit_b_16':
        model = models.vit_b_16(weights=weights)
        model.heads.head = nn.Linear(model.heads.head.in_features, num_classes)
    elif model_name == 'regnet_y_400mf':
        model = models.regnet_y_400mf(weights=weights)
        model.stem[0] = nn.Conv2d(1, 32, kernel_size=3, stride=2, padding=1, bias=False)
        model.fc = nn.Linear(model.fc.in_features, num_classes)
    elif model_name == 'convnext_tiny':
        model = models.convnext_tiny(weights=weights)
        model.features[0][0] = nn.Conv2d(1, 96, kernel_size=4, stride=4, bias=False)
        model.classifier[2] = nn.Linear(model.classifier[2].in_features, num_classes)
    elif model_name == 'openl3':
        model = torchopenl3.models.load_audio_embedding_model(input_repr="mel256", content_type="music", embedding_size=512, pretrained=pretrained)
        model.fc = nn.Linear(model.fc.in_features, num_classes)
    elif model_name == 'soundnet':
        model = SoundNet8_pytorch(num_classes=num_classes)
    elif model_name == 'wavenet':
        model = WaveNet(num_classes=num_classes)
    elif model_name == 'mfcc_cnn':
        model = MFCCCNN(num_classes=num_classes)
    else:
        raise ValueError("Unsupported model type")
    
    return model

class SoundNet8_pytorch(nn.Module):
    def __init__(self, num_classes):
        super(SoundNet8_pytorch, self).__init__()
        self.layers = nn.Sequential(
            nn.Conv1d(1, 16, kernel_size=64, stride=2, padding=32),
            nn.BatchNorm1d(16),
            nn.ReLU(),
            nn.MaxPool1d(8),
            nn.Conv1d(16, 32, kernel_size=32, stride=2, padding=16),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.MaxPool1d(8),
            nn.Conv1d(32, 64, kernel_size=16, stride=2, padding=8),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.MaxPool1d(8),
            nn.Conv1d(64, 128, kernel_size=8, stride=2, padding=4),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.MaxPool1d(8),
            nn.Conv1d(128, 256, kernel_size=4, stride=2, padding=2),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.MaxPool1d(8),
            nn.Conv1d(256, 512, kernel_size=4, stride=2, padding=2),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.MaxPool1d(8),
            nn.Conv1d(512, 1024, kernel_size=4, stride=2, padding=2),
            nn.BatchNorm1d(1024),
            nn.ReLU(),
            nn.MaxPool1d(8),
        )
        self.fc = nn.Linear(1024, num_classes)
        
    def forward(self, x):
        x = self.layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x


class WaveNet(nn.Module):
    def __init__(self, num_classes):
        super(WaveNet, self).__init__()
        self.layers = nn.Sequential(
            nn.Conv1d(1, 32, kernel_size=2, dilation=1),
            nn.ReLU(),
            nn.Conv1d(32, 64, kernel_size=2, dilation=2),
            nn.ReLU(),
            nn.Conv1d(64, 128, kernel_size=2, dilation=4),
            nn.ReLU(),
            nn.Conv1d(128, 256, kernel_size=2, dilation=8),
            nn.ReLU(),
            nn.Conv1d(256, 512, kernel_size=2, dilation=16),
            nn.ReLU(),
            nn.AdaptiveAvgPool1d(1)
        )
        self.fc = nn.Linear(512, num_classes)
        
    def forward(self, x):
        x = self.layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x


class MFCCCNN(nn.Module):
    def __init__(self, num_classes=3, dropout_rate=0.5):
        super(MFCCCNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1, stride=1)
        self.pool1 = nn.MaxPool2d(kernel_size=(2, 1), stride=(2, 1))
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1, stride=1)
        self.pool2 = nn.MaxPool2d(kernel_size=(2, 1), stride=(2, 1))
        self.dropout1 = nn.Dropout(p=dropout_rate)
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1, stride=1)
        self.pool3 = nn.MaxPool2d(kernel_size=(2, 1), stride=(2, 1))
        self.dropout2 = nn.Dropout(p=dropout_rate)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(self._get_conv_output_size(), 128)
        self.dropout3 = nn.Dropout(p=dropout_rate)
        self.fc2 = nn.Linear(in_features=128, out_features=num_classes)

    def _get_conv_output_size(self):
        # Create a dummy input with the same dimensions as a typical input
        dummy_input = torch.zeros(1, 1, 224, 224)
        dummy_output = self._forward_conv_layers(dummy_input)
        return dummy_output.view(1, -1).size(1)

    def _forward_conv_layers(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool1(x)
        x = F.relu(self.conv2(x))
        x = self.pool2(x)
        x = self.dropout1(x)
        x = F.relu(self.conv3(x))
        x = self.pool3(x)
        x = self.dropout2(x)
        return x

    def forward(self, x):
        x = self._forward_conv_layers(x)
        x = self.flatten(x)
        x = F.relu(self.fc1(x))
        x = self.dropout3(x)
        logits = self.fc2(x)
        return logits
```

```python
import os
import torch

# Preprocess input to adjust dimensions
def preprocess_input(x):
    # Assuming x is of shape [batch_size, 1, height, width]
    if x.size(2) < 224 or x.size(3) < 224:
        x = F.pad(x, (0, max(0, 224 - x.size(3)), 0, max(0, 224 - x.size(2))))
    x = F.interpolate(x, size=(224, 224), mode='bilinear', align_corners=False)
    return x


def train_model(model, dataloaders, criterion, optimizer, device, model_name, model_path, num_epochs=25, scheduler=None):
    train_loss_history = []
    val_loss_history = []
    train_acc_history = []
    val_acc_history = []
    best_acc = 0.0

    # Ensure the directory to save model checkpoints exists
    os.makedirs(model_path, exist_ok=True)

    # Initialize mixed precision training
    scaler = GradScaler()

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        corrects = 0
        
        for inputs, labels in dataloaders['train']:
            inputs, labels = inputs.to(device), labels.to(device)
            inputs = preprocess_input(inputs)  # Apply preprocessing here

            optimizer.zero_grad()
            # Mixed Precision Training - to speed up.
            with autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        
            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            corrects += torch.sum(preds == labels.data)
        
        epoch_loss = running_loss / len(dataloaders['train'].dataset)
        epoch_acc = corrects.double() / len(dataloaders['train'].dataset)
        train_loss_history.append(epoch_loss)
        train_acc_history.append(epoch_acc.item())

        print(f'Epoch {epoch+1}/{num_epochs} - Train Loss: {epoch_loss:.4f} - Train Acc: {epoch_acc:.4f}')
        
        # Validation step
        model.eval()
        val_loss = 0.0
        corrects = 0
        
        with torch.no_grad():
            for inputs, labels in dataloaders['val']:
                inputs, labels = inputs.to(device), labels.to(device)
                inputs = preprocess_input(inputs)  # Apply preprocessing here
                
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                corrects += torch.sum(preds == labels.data)
        
        val_loss /= len(dataloaders['val'].dataset)
        val_acc = corrects.double() / len(dataloaders['val'].dataset)
        val_loss_history.append(val_loss)
        val_acc_history.append(val_acc.item())

        print(f'Epoch {epoch+1}/{num_epochs} - Val Loss: {val_loss:.4f} - Val Acc: {val_acc:.4f}')
        
        # Save checkpoint if validation accuracy is the best we've seen so far
        if val_acc > best_acc:
            print('Saving checkpoint...')
            state = {
                'model': model.state_dict(),
                'acc': val_acc,
                'epoch': epoch,
            }
            checkpoint_path = os.path.join(model_path, f'{model_name}_ckpt.pth')
            torch.save(state, checkpoint_path)
            best_acc = val_acc

        if scheduler:
            scheduler.step()
    
    return train_loss_history, val_loss_history, train_acc_history, val_acc_history
```
